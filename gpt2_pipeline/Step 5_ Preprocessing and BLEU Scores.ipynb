```python
import pandas as pd

def process_row(row):
    # Clean Recipe Title: Remove newline characters
    recipe_title = row['Recipe Title'].replace('\n', '').strip()

    # Clean Recipe Instructions: Truncate text after "Voila Enjoy your recipe :)"
    recipe_instructions = row['Recipe Instructions']
    if "Voila Enjoy your recipe :)" in recipe_instructions:
        recipe_instructions = recipe_instructions[:recipe_instructions.index("Voila Enjoy your recipe :)")].strip()

    return {
        'Random Ingredients': row['Random Ingredients'],  # Assuming other columns are correctly formatted
        'Recipe Title': recipe_title,
        'Ingredient Phrases': row['Ingredient Phrases'],  # Assuming correct formatting
        'Recipe Instructions': recipe_instructions
    }

# Assuming combined_recipes is already your DataFrame loaded with the raw data
cleaned_data = [process_row(row) for index, row in combined_recipes.iterrows()]
clean_recipes_df = pd.DataFrame(cleaned_data)

# Check the first few entries to ensure proper formatting
print(clean_recipes_df.head())
```

```python
nltk.download('punkt')
nltk.download('wordnet')

def preprocess_text(text):
    return nltk.word_tokenize(text.lower())

# Prepare data - combine all text fields into a single string for each recipe and then tokenize
reference_texts = [[preprocess_text(f"{row['Recipe Title']} {row['Ingredient Phrases']} {row['Recipe Instructions']}")] for index, row in recipes_df.iterrows()]
candidate_texts = [preprocess_text(f"{row['Recipe Title']} {row['Ingredient Phrases']} {row['Recipe Instructions']}") for index, row in clean_recipes_df.iterrows()]

# Calculate BLEU Score
bleu_score = corpus_bleu(reference_texts, candidate_texts)

# Calculate METEOR Score
meteor_scores = [meteor_score(ref, cand) for ref, cand in zip(reference_texts, candidate_texts)]
average_meteor = sum(meteor_scores) / len(meteor_scores)

# Calculate BERTScore
P, R, F1 = score([" ".join(cand) for cand in candidate_texts], [" ".join(ref[0]) for ref in reference_texts], lang="en", verbose=True)
bertscore_f1 = F1.mean().item()

# Prepare ROUGE scorer
rouge_scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': []}

for ref, cand in zip(reference_texts, candidate_texts):
    scores = rouge_scorer.score(' '.join(ref[0]), ' '.join(cand))
    rouge_scores['rouge1'].append(scores['rouge1'].fmeasure)
    rouge_scores['rouge2'].append(scores['rouge2'].fmeasure)
    rouge_scores['rougeL'].append(scores['rougeL'].fmeasure)

# Average ROUGE scores
rouge_scores = {key: sum(values) / len(values) for key, values in rouge_scores.items()}

# Output results
print("BLEU Score:", bleu_score)
print("Average METEOR Score:", average_meteor)
print("Average BERTScore F1:", bertscore_f1)
print("Average ROUGE Scores:", rouge_scores)
```
